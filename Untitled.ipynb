```python
# create a dist plot
f, axes = plt.subplots(ncols=3, figsize=(15, 6))

sns.distplot(exam_df['reading score'], color='skyblue', ax=axes[0])
sns.distplot(exam_df['math score'], color='gold', ax=axes[1])
sns.distplot(exam_df['writing score'], color='red', ax=axes[2])

f.tight_layout()
```

```python
-female students are getting more marks on reading and write but male students are getting more on math
-lunch is affecting student notes - students with free/reduced lunch get worse marks
-also students with parents have master degree get better marks
-lastly test preparation course is another important dimension - sudents who has completed this course get better marks
```
